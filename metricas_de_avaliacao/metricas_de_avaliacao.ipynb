## **Desafio de Projeto** <br>Cálculo de Métricas de Avaliação de Aprendizado
<p> Este desafio de projeto se refere ao módulo "Teoria do Aprendizado Estatístico" do bootcamp "BairesDev Machine Learning Training", na plataforma DIO.
<p> O projeto consiste em obter, a partir de um modelo de aprendizado de máquina, as seguintes métricas de desempenho:


*   Acurácia
*   Precisão
*   Sensibilidade
*   Especificidade
*   F-Score


<p>Para um modelo de classificação com duas clases, as métricas são obtidas a partir da tabela verdade do modelo, a qual contém os valores a seguir:


*   Verdadeiros Positivos: quantidade de amostras da classe 1 rotuladas pelo modelo com a classe 1;
*   Falsos Positivos: quantidade de amostras da classe 2 rotuladas pelo modelo com a classe 1;
*   Falsos Negativos: quantidade de amostras da classe 1 rotuladas pelo modelo com a classe 2;
*   Verdadeiros Negativos: quantidade de amostras da classe 2 rotuladas pelo modelo com a classe 2.






```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import RidgeClassifier

print('Tudo certo com as importações')
```

## Geração de Dataset
<p> Para fins didáticos, foi gerado um dataset com 300 amostras divididas em duas classes, utilizando a função `make_classification`, disponível na biblioteca `sklearn`.

```python
# Geração do dataset
x_data, y_data = make_classification(
    n_samples=300,
    n_classes=2,
    n_features=2,
    n_redundant=0,
    random_state=42,
    n_informative=1,
    n_clusters_per_class=1)
plt.figure(figsize=(10,5))
plt.scatter(x_data[:, 0], x_data[:, 1], c=y_data)
plt.title('Dataset utilizado')
plt.show()
```

```python
print(x_data.shape, y_data.shape)
```

## Definição do modelo
<p>Para o problema de classificação, foi utilizado o modelo `RidgeClassifier`, da biblioteca `sklearn`.
<p>O modelo foi treinado com todas as amostras do dataset gerado anteriormente.

```python
model = RidgeClassifier()
model.fit(x_data, y_data)
```

## Avaliação do modelo
<p> A partir da matriz de confusão do modelo, foram calculadas as métricas de desempenho, como segue.

```python
# Matriz de confusão:
tn, fp, fn, tp = confusion_matrix(y_data, model.predict(x_data)).ravel()
print('Verdadeiros Positivos:', tp)
print('Falsos Positivos:', fp)
print('Falsos Negativos:', fn)
print('Verdadeiros Negativos:', tn)
print()
print("Métricas do modelo:")
accuracy = (tp+tn)/(tp+fp+fn+tn)
precision = (tp)/(tp+fp)
recall = (tp)/(tp+fn)
specificity = (tn)/(fp+tn)
f_score = 2 * precision * recall / (precision + recall)
print('Acurácia:', accuracy)
print('Precisão:', precision)
print('Sensibilidade:', recall)
print('Especificidade:', specificity)
print('F-score:', f_score)
```

##Avaliação
<p>Para os dados de treino, obtemos uma acuária de 96% e uma sensibilidade de 92,71%, o que indica que, a princípio, o modelo não apresenta overfit e é razoavelmente intolerante a falsos positivos.
<p>Como média harmônica, temos um f-score de 95,89%.

##Considerações finais
<p>É importante observar que as métricas foram calculadas a partir da predição do modelo tendo como entrada os dados de treino, o que foi feito apenas para fins didáticos.
<p>Em um cenário real, os dados disponíveis devem ser divididos em treino, validação e teste, para evitar o enviesamento excessivo do modelo e estimular sua capacidade de generalização para amostras desconhecidas.
