## Desafio de Projeto: Redução de Dimensionalidade de Imagens
<p>Este desafio faz parte do Bootcamp "BairesDev - Machine Learning Training", da DIO.
<p>O desafio consiste em, a partir de uma umagem colorida, gerar duas novas imagens, sendo a primeira em escala de cinza e a segunda em preto e branco. Porém, não se pode utilizar bibliotecas prontas para processamento de imagens.

```python
import numpy as np # para trabalhar as imagens como arrays
import matplotlib.pyplot as plt # para exibir as imagens
from PIL import Image #para conversão de imagem para array e vice-versa
```

### 1ª Etapa: Carregar a imagem RGB e convertê-la em matriz de cores

```python
img_path = 'lena.jpg'
def load_img(img_path):
    img = Image.open(img_path)
    return np.asarray(img)

rgb_img = load_img(img_path)
plt.imshow(rgb_img)
plt.yticks([])
plt.xticks([])
plt.show()
```

### 2ª Etapa: Converter imagem de RGB para escala de cinza


```python
def image_rgb_to_gray(img):
    return np.mean(img, axis=2)


gray_img = image_rgb_to_gray(rgb_img)
show_gray_img = Image.fromarray(gray_img)
plt.imshow(show_gray_img)
plt.yticks([])
plt.xticks([])
plt.show()
```

### 3ª Etapa: Converter pixel de escala de cinza para preto e branco

```python
# Função para converter pixels de escala de cinza para preto e branco   
def image_gray_to_bw(img, treshold=90):
    return (img >= treshold) * 255

bw_img = image_gray_to_bw(gray_img)
show_bw_img = Image.fromarray(bw_img)
plt.imshow(show_bw_img)
plt.yticks([])
plt.xticks([])
plt.show()
```
